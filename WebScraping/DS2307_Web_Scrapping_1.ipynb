# 3) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape and make data frame 
#a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
#b) Top 10 ODI Batsmen along with the records of their team andrating.
#c) Top 10 ODI bowlers along with the records of their team andrating.

# ANSWER 3a):

```python
page_a=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
page_a
```

```python
soup = BeautifulSoup(page_a.text, "html.parser")
soup
```

```python
ODI_Teams = []

for team in soup.select("tbody tr"):
    countries = team.select('span',class_="u-hide-phablet")[1].text
    matches = team.select("td")[2].text
    points = team.select("td")[3].text
    ratings = team.select("td")[4].text.strip()
    ODI_Teams.append((countries, matches, points, ratings))

df = pd.DataFrame(ODI_Teams, columns=["Team", "Matches", "Points", "Ratings"])
df = df.head(10)  
print(df)
```

```python
soup=BeautifulSoup(page_a.content)
soup
```

```python
ODI_Teams = []

for team in soup.select("tbody tr"):
    countries = team.select('span',class_="u-hide-phablet")[1].text
    matches = team.select('td',class_='table-body__cell u-center-text')[2].text
    points = team.select("td")[3].text
    ratings = team.select('td',class_='table-body__cell u-text-right rating')[4].text.strip()
    ODI_Teams.append((countries, matches, points, ratings))

df = pd.DataFrame(ODI_Teams, columns=["Team", "Matches", "Points", "Ratings"])
df = df.head(10)  
print(df)
```

```python
ODI_Women = []

for team_row in soup.select("tr.table-body"):
    name = team_row.select('.table-body__cell.rankings-table__name')[0].get_text().strip()
    team = team_row.select('.table-body__cell.rankings-table__team')[0].get_text().strip()
    ratings = team_row.select('.table-body__cell.rating')[0].get_text().strip()
    ODI_Women.append((name, team, ratings))

df = pd.DataFrame(ODI_Women, columns=["Names", "Team", "Ratings"])
df = df.head(10)
print(df)
```

# 5) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame
i) Headline
ii) Time
iii) News Link

# ANSWER 5i)

```python
page_i=requests.get('https://www.cnbc.com/world/?region=world')
page_i
```

```python
soup=BeautifulSoup(page_i.content)
soup
```

```python
Headline=[]
for i in soup.find_all('div',class_='RiverHeadline-headline RiverHeadline-hasThumbnail'):
   Headline.append(i.text)
Headline
```

# ANSWER 5ii)

```python
Time=[]
for i in soup.find_all('time'):
   Time.append(i.text)
Time
```

# ANSWER 5iii)

```python
News_links = []

for link in soup.find_all('a'):
    href = link.get('href')
    if href and href.startswith('/'):
        full_link = "https://www.cnbc.com{href}"
        News_links.append(full_link)

df = pd.DataFrame(News_links, columns=["News Links"])
print(df)
```

# 7) Write a python program to scrape mentioned details from dineout.co.inand make data frame
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
page_7=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page_7
```

```python
soup=BeautifulSoup(page_7.content,"html.parser")
soup
```

```python
Restaurant_Name=[]
for i in soup.find_all('div',class_="restnt-info cursor"):
    Restaurant_Name.append(i.text)

Restaurant_Name
```

```python
price=[]
for i in soup.find_all('span',class_="double-line-ellipsis"):
    price.append(i.text)
    
price
```

```python
cuisines = []

for i in soup.find_all('span', class_="double-line-ellipsis"):
    text = i.text.replace('₹', '')
    price_cuisine = text.split('|',1)[1]
    cuisines.append(price_cuisine.strip())
print(cuisines)
```

```python
Location=[]
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    Location.append(i.text)
Location
```

```python
Ratings=[]
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    Ratings.append(i.text)
Ratings
```

```python
image_url=[]
for i in soup.find_all('img',class_="no-img"):
    image_url.append(i['data-src'])
image_url
```

```python
df=pd.DataFrame({"Restaurant Name":Restaurant_Name, "Cusines":cuisines, "Location":Location, "Ratings":Ratings,"Images_Url":image_url})
df
```
